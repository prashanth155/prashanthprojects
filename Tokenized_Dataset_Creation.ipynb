```python
import numpy as np


def train_test_val(df, train_percent = 0.6,val_percent = 0.2, seed = None):
  np.random.seed(seed)
  perm = np.random.permutation(df.index)

  m = len(df.index)
  train_end = int(train_percent*m)
  val_end = int(val_percent*m) + train_end

  train = df.iloc[perm[:train_end]]
  vali = df.iloc[perm[train_end:val_end]]
  test = df.iloc[perm[val_end:]]

  return train, vali, test
```

```python
df = pd.read_csv('/content/drive/MyDrive/FullText.csv', index_col = False)
```

```python
train, validate, test = train_test_val(df)

train
```

```python
train = train.drop(columns = ['Unnamed: 0'], axis = 1)
train = train.reset_index()
train = train.drop(columns = ['index'], axis = 1)
```

```python
from transformers import pipeline, set_seed
from datasets import load_dataset, load_from_disk
import matplotlib.pyplot as plt
from datasets import load_dataset
import pandas as pd
from datasets import load_dataset, load_metric

from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

import nltk
from nltk.tokenize import sent_tokenize

from tqdm import tqdm
import torch

nltk.download("punkt")
```

```python
model_ckpt = "google/pegasus-cnn_dailymail"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
```

```python
train_lis = []
train_sum = []

import warnings
warnings.filterwarnings("ignore")

for i in range(len(train.index)):

  train_lis.append(tokenizer(train.News.iloc[i] , max_length = 1024, truncation = True ))
  with tokenizer.as_target_tokenizer():
    train_sum.append(tokenizer(train.Summary.iloc[i], max_length = 128, truncation = True ))
```

```python
val_lis = []
val_sum = []

for i in range(len(validate.index)):

  val_lis.append(tokenizer(validate.News.iloc[i] , max_length = 1024, truncation = True ))
  with tokenizer.as_target_tokenizer():
    val_sum.append(tokenizer(validate.Summary.iloc[i], max_length = 128, truncation = True ))
```

```python
test_lis = []
test_sum = []

for i in range(len(test.index)):

  test_lis.append(tokenizer(test.News.iloc[i] , max_length = 1024, truncation = True ))
  with tokenizer.as_target_tokenizer():
    test_sum.append(tokenizer(test.Summary.iloc[i], max_length = 128, truncation = True ))
```

```python
final_train = pd.DataFrame(columns = ['id', 'dialogue', 'summary', 'input_ids', 'attention_mask', 'labels'])
final_test = pd.DataFrame(columns = ['id', 'dialogue', 'summary', 'input_ids', 'attention_mask', 'labels'])
final_validate = pd.DataFrame(columns = ['id', 'dialogue', 'summary', 'input_ids', 'attention_mask', 'labels'])
```

```python
for i in range(len(train.index)):

  final_train.loc[i] = [i,train.News.iloc[i],train.Summary.iloc[i], train_lis[i]['input_ids'],train_lis[i]['attention_mask'],train_sum[i]['input_ids']]

```

```python
for i in range(len(test.index)):

  final_test.loc[i] = [i,test.News.iloc[i],test.Summary.iloc[i], test_lis[i]['input_ids'],test_lis[i]['attention_mask'],test_sum[i]['input_ids']]
```

```python
for i in range(len(validate.index)):

  final_validate.loc[i] = [i,validate.News.iloc[i],validate.Summary.iloc[i], val_lis[i]['input_ids'], val_lis[i]['attention_mask'], val_sum[i]['input_ids']]
```
